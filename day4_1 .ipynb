```python
import numpy as np
```

```python
path = "Data/day4_1.txt"
X = []

with open(path, "r", encoding="utf-8") as f:
    for line in f:
        # Convert string to list of characters for 2D array
        X.append(list(line.strip()))

X = np.array(X)
print(f"Shape: {X.shape}")
print(X)
```

```python
def check_adjacent_cells(row, col):
    count = 0
    rows, cols = X.shape
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue  # Skip the cell itself
            
            ni, nj = row + i, col + j
            
            # Check boundaries
            if 0 <= ni < rows and 0 <= nj < cols:
                if X[ni,nj] == "@":
                    count += 1
                
    # print(f"Cell ({row}, {col}) has {count} adjacent '@' cells.")        
    return 1 if count < 4 else 0
```

```python
total = 0
Y = X.copy()
#Y = X[:1, :10]
for row in range(len(Y)):
    for col in range(len(Y[0])):
        
        if Y[row][col] == ".":
            continue
        else:
            total +=check_adjacent_cells(row, col)

print(total)
```
