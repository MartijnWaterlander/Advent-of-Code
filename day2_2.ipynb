```python
import numpy as np
import pytest
```

```python
path = "Data/day2_1.txt"
with open(path, "r", encoding="utf-8") as f:
    data = f.read()

data_strip = str(data)[:-1]
data_strip = [p.strip() for p in data_strip.split(",")]
```

```python
tuples = []
for item in data_strip:
    lower, higher = item.split("-")
    tuples.append((int(lower), int(higher)))

total = 0
```

```python
def sequence_detector(tuple):
    """
    Detects numbers in a range that are formed by repeating a sequence twice (e.g. 123123).
    Updates the global 'total'.

    Args:
        tuple (tuple): Range (lower, higher).
    """
    lower, higher = tuple
    
    for number in range(lower, higher+1):
        str_number = str(number)
        length = len(str_number)
        
        # continue if the length is uneven
        if length % 2 != 0:
            continue

        first_half = int(str_number[:length//2])
        second_half = int(str_number[length//2:])

        if first_half == second_half:
            global total
            total += int(str_number)
```

```python
def is_invalid(n: int) -> bool:
    """
    Checks if a number is formed by repeating the same sequence twice.

    Args:
        n (int): The number to check.

    Returns:
        bool: True if the number is a repeated sequence, False otherwise.
    """
    s = str(n)
    if len(s) % 2 != 0:
        return False
    half = len(s) // 2
    return int(s[:half]) == int(s[half:])

def find_invalid_ids(lower: int, higher: int):
    """
    Finds all invalid IDs in a given range.

    Args:
        lower (int): Lower bound.
        higher (int): Upper bound.

    Returns:
        list: List of invalid IDs.
    """
    return [n for n in range(lower, higher + 1) if is_invalid(n)]

@pytest.mark.parametrize("range_tuple,expected", [
    ((11, 22), [11, 22]),
    ((95, 115), [99]),
    ((998, 1012), [1010]),
    ((1188511880, 1188511890), [1188511885]),
    ((222220, 222224), [222222]),
    ((1698522, 1698528), []),
    ((446443, 446449), [446446]),
    ((38593856, 38593862), [38593859]),
])
def test_edge_cases(range_tuple, expected):
    lower, higher = range_tuple
    assert find_invalid_ids(lower, higher) == expected
```

```python
for i in range(len(tuples)):
    sequence_detector(tuples[i])

print(total)
```
