```python
import numpy as np

path = "Data/day6_1.txt"
X = []
ORIGINAL = []
with open(path, "r", encoding="utf-8") as f:
    for line in f:
        ORIGINAL.append(line)
        line = line.split()
        print(line)
        X.append(line)

X = np.array(X)

# Swap first and last row
Y = np.array(X)
Y[0,:] = X[-1,:]
Y[-1,:] = X[0,:]
Y

```

```python
ORIGINAL = np.array(ORIGINAL)
ORIGINAL[4][0]

sided = []
for i in range(len(ORIGINAL[4])):
    if ORIGINAL[4][i] == " ":
        continue
    elif ORIGINAL[4][i] == "+" or ORIGINAL[4][i] == "*":
        for j in range(0, 4):
            if ORIGINAL[j][i] == " ":
                sided.append("R")
                break
            else:
                sided.append("L")
                break
sided        
```

```python
def calculate_temp(temp, sign):
    if sign == "+":
        total = 0
        for k in range(len(temp)):
            total += int(temp[k])
        return total
    elif sign == "*":
        total = 1
        for k in range(len(temp)):
            total *= int(temp[k])
        return total
```

```python
# Loop over columns
total = 0
temp = []
for i in range(len(ORIGINAL[0])):
    row = ""
    if ORIGINAL[4][i] == "*":
        sign = "*"
    elif ORIGINAL[4][i] == "+":
        sign = "+"    

    # Loop over rows
    for j in range(len(ORIGINAL) -1):
        row+=(ORIGINAL[j][i])

    # Check if row is not empty
    if row.strip() == "":
        print(f"Calculating temp for {temp} with sign {sign}")
        total += calculate_temp(temp, sign)
        temp = []
        continue  
    temp.append(row.strip())
    
total += calculate_temp(temp, sign)       
        
total
```

```python
Z = np.array(Y)
for i in range(Z.shape[1]):
    length = max(len(Y[j,i]) for j in range(1, Y.shape[0]))
    if Z[0,i] == '*':
        for j in range(1, Z.shape[0]):
            val = Z[j,i]
            len_val = len(val)
            add = length - len_val

            if sided[i] == 'L':
                Z[j,i] = val+ '1'*add 
            else:
                Z[j,i] = '1'*add + val

    elif Z[0,i] == '+':
        for j in range(1, Z.shape[0]):
            val = Z[j,i]
            len_val = len(val)
            add = length - len_val
            if sided[i] == 'L':
                Z[j,i] = val+ '0'*add 
            else:
                Z[j,i] = '0'*add + val

Z
```

```python
total = 0

for i in range(Z.shape[1]):
    nums = []
    for j in range(Z.shape[0]):
        nums.append(Z[j,i])

    if nums[0] == "*":
        
        for k in range(max(len(Z[j,i]) for j in range(1, Z.shape[0]))):

            sum = int(nums[1][k]) * int(nums[2][k]) * int(nums[3][k]) * int(nums[4][k])
            print(f"Product of {nums[1][k]} * {nums[2][k]} * {nums[3][k]} * {nums[4][k]}  = {sum}")
            total += sum
    
    elif nums[0] == "+":
       
       for k in range(max(len(Z[j,i]) for j in range(1, Z.shape[0]))):

            sum = int(nums[1][k]) + int(nums[2][k]) + int(nums[3][k]) + int(nums[4][k])
            print(f"Sum of {nums} is {sum}")
            total += sum
    
    
print(total)
```
