```python
path = "Data/day3_1.txt"

battery_voltages = 0
with open(path, "r", encoding="utf-8") as f:
    for lineno, line in enumerate(f, start=1):
        highest = highest_voltage(line)
        battery_voltages += highest

print(battery_voltages)
```

```python
def highest_voltage(line):
    """
    Finds the highest voltage combination from two digits in the line.

    Args:
        line (str): The input string of digits.

    Returns:
        int: The highest calculated voltage.
    """
    highest = 0
    line = line.strip()
    for i in range(len(line)):
        #print(line[i])
        for j in range(i+1, len(line)):
            #print(line[j])
            combination = int(line[i] + line[j])
            if combination > highest:
                highest = combination
    return highest
```
