```python
def highest_voltage(line):
    """
    Calculates the highest voltage by selecting digits from the line.

    Args:
        line (str): The input string of digits.

    Returns:
        int: The highest voltage formed.
    """
    highest_index = 0
    highest = ''
    line = line.strip()
    length = 12
    for i in range(12):
        
        highest_num = 0

        if length == 1:

            start = line[highest_index:]

        else:
            start = line[highest_index:-(length-1)]

        for idx, char in enumerate(start):
            
            if int(char) > highest_num:
                highest_num = int(char)
                highest_index = idx

        highest_index += 1       
        highest += str(highest_num)
        highest_num = 0
        line = line[highest_index:]
        highest_index = 0
        length -= 1
    return int(highest)
```

```python
path = "Data/day3_1.txt"

battery_voltages = 0
with open(path, "r", encoding="utf-8") as f:
    for lineno, line in enumerate(f, start=1):
        highest = highest_voltage(line)
        battery_voltages += highest

print(battery_voltages)
```
