```python
import numpy as np

path = "Data/day6_1.txt"
X = []
with open(path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.split()
        print(line)
        X.append(line)

X = np.array(X)
```

```python
# Swap first and last row
Y = np.array(X)
Y[0,:] = X[-1,:]
Y[-1,:] = X[0,:]
Y
```

```python
total = 0

for i in range(Y.shape[1]):
    nums = []
    for j in range(Y.shape[0]):
        nums.append(Y[j,i])

    if nums[0] == "*":
        sum = int(nums[1]) * int(nums[2]) * int(nums[3]) * int(nums[4])
        print(f"Sum of {nums} is {sum}")
    elif nums[0] == "+":
        sum = int(nums[1]) + int(nums[2]) + int(nums[3]) + int(nums[4])
        print(f"Sum of {nums} is {sum}")
    total += sum
    
print(total)
```
