```python
path = "Data/day2_1.txt"
with open(path, "r", encoding="utf-8") as f:
    data = f.read()

data_strip = str(data)[:-1]
data_strip = [p.strip() for p in data_strip.split(",")]
```

```python
tuples = []
for item in data_strip:
    lower, higher = item.split("-")
    tuples.append((int(lower), int(higher)))
```

```python
def sequence_detector(tuple):
    """
    Detects if numbers in a range (defined by tuple) are composed of repeating sequences.
    Updates the global 'total' with the sum of such numbers.

    Args:
        tuple (tuple): A tuple (lower, higher) defining the range.
    """
    lower, higher = tuple
    
    for number in range(lower, higher+1):
        str_number = str(number)
        length = len(str_number)
        
        for i in range(2, length+1):
            
            if length % i != 0:
                continue
            
            seq_len = length//i

            seqs = []
            for j in range(0, i):
                segment = str_number[j*seq_len:(j+1)*seq_len]
                seqs.append(segment)

            if all(x == seqs[0] for x in seqs):
                global total
                total += number
                print(f"Found sequence in number: {number}")
                break
                
                

        # first_half = int(str_number[:length//2])
        # second_half = int(str_number[length//2:])

        # if first_half == second_half:
        #     global total
        #     total += int(str_number)
```

```python
total = 0

for i in range(len(tuples)):
    sequence_detector(tuples[i])

print(total)
```
