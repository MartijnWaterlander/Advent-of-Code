```python
path = "Data/day5_1.txt"
IDS = []

with open(path, "r", encoding="utf-8") as f:
    for line in f:
        IDS.append(line.strip())

ID_ranges = IDS[:177]
IDS = IDS[178:]
```

```python
def is_between(id1, id2,id):
    global fresh
    if int(id1) <= int(id) <= int(id2):
        fresh += 1
        return True
    return False
    
```

```python
ranges = []

for id_range in ID_ranges:
    id1, id2 = id_range.split('-')
    ranges.append((int(id1), int(id2)))

fresh = 0

for i in range(0, len(IDS)):
    for j in range(0, len(ranges)):
        id1, id2 = ranges[j]
        if is_between(id1, id2, IDS[i]):
            break
            
    
print(fresh)
           
```
